```python
from yolov3_core import *
from yolov3_tracker_latest import *
```

```python
# create settings dictionary
settings = {'model_def': 'cfg/yolov3-spp-1cls.cfg',
            'weights_path': 'weights/best.pt',
            'class_path': 'old_608/classes.names',
            'img_size': 608,
            'iou_thres': 0.6,
            'no_gpu': True,
            'conf_thres': 0.3,
            'batch_size': 1,
            'augment': False,
            'classes': None}

# initiate model
Yolo = YoloModelLatest(settings)
```

```python
# load image
img_path = "data/samples/exp506_random3.png"
img = cv2.imread(img_path)
# process and pass image through model
processed_images = ImageProcessor(img)
model_input = processed_images.image_slices
outputs = Yolo.pass_model(model_input)
```

```python
fig=plt.figure(figsize=(15, 15))
for output in outputs:
    x1, y1, x2, y2, conf, cls_conf = output
    cv2.rectangle(img, (x1,y1), (x2,y2), (255,100,140),2)
    pass

plt.imshow(img)
    
```
